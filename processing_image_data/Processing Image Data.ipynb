Converting RGB images to grayscale image using OpenCV

```python
!pip install opencv-python
```

```python
import cv2
```

```python
img = cv2.imread("dog.jpg")
```

```python
type(img)
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
grayscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
grayscale_img.shape
```

```python
cv2.imshow(grayscale_img)
```

```python
cv2.imwrite("dog_grayscale.jpg",grayscale_img)
```
